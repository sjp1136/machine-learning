{'max_features': 700, 'n_estimators': 1200}

```
from sklearn.pipeline import Pipeline
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn import preprocessing

url = 'https://raw.githubusercontent.com/davio606/CS-4774-data/master/data.csv'
num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('std_scaler', StandardScaler()),
    ])

num_pipeline_y = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
    ])

def new_clean (data_in):

  data11 = data_in
  data12 = data11.append(pd.Series(), ignore_index=True)
  data13 = data12


  data13.drop(data13.loc[data13['Make'] == "Ferrari"].index, inplace=True)
  data13.drop(data13.loc[data13['Make'] == "McLaren"].index, inplace=True)
  data13.drop(data13.loc[data13['Make'] == "Bentley"].index, inplace=True)
  data13.drop(data13.loc[data13['Make'] == "Lamborghini"].index, inplace=True)
  data13.drop(data13.loc[data13['Make'] == "Aston Martin"].index, inplace=True)
  data13.drop(data13.loc[data13['Make'] == "Bugatti"].index, inplace=True)
  data13.drop(data13.loc[data13['Make'] == "Maybach"].index, inplace=True)
  data13.drop(data13.loc[data13['Make'] == "Spyker"].index, inplace=True)
  data13.drop(data13.loc[data13['Make'] == "Rolls-Royce"].index, inplace=True)
  data13.drop(data13.loc[data13['Make'] == "Maserati"].index, inplace=True)



  cat_data14 = data13[["Make","Model","Engine Fuel Type","Transmission Type","Driven_Wheels","Market Category","Vehicle Size","Vehicle Style"]]
  num_data14 = data13.drop(["Make","Model","Engine Fuel Type","Transmission Type","Driven_Wheels","Market Category","Vehicle Size","Vehicle Style"], axis=1)
  cat_data15 = cat_data14.replace(np.nan, '', regex=True)
  num_X = num_data14.drop(["MSRP"], axis=1)
  num_y = num_data14[["MSRP"]]

  cat_encoder14 = OneHotEncoder(sparse=False)
  cat_1hot14 = cat_encoder14.fit_transform(cat_data15)
  cat_1hot14

  num_data15 = num_pipeline.fit_transform(num_X)
  num_yp = num_pipeline_y.fit_transform(num_y)

  clean_data2 = np.append(cat_1hot14, num_data15, 1)
  clean_data3 = np.append(clean_data2, num_yp, 1)

  return clean_data3

data10 = pd.read_csv(url)
clean_data3 = new_clean(data10)
```

```
train_set3, test_set3 = train_test_split(clean_data3, test_size=0.1, random_state=42)
train_setNN3, valid_set3 = train_test_split(train_set3, test_size=0.1, random_state=42)
data_length = len(clean_data3[0])-1
X_train3, y_train3 = train_set3[:,0:data_length].copy(), train_set3[:,data_length].copy()
X_trainNN3, y_trainNN3 = train_setNN3[:,0:data_length].copy(), train_setNN3[:,data_length].copy()
X_valid3, y_valid3 = valid_set3[:,0:data_length].copy(), valid_set3[:,data_length].copy()
X_test3, y_test3 = test_set3[:,0:data_length].copy(), test_set3[:,data_length].copy()
```
